## The Math of PCA ###

We start with PCA in 2D, which is convinient because we can plot everything. Performing PCA in 2D corrsponds to identifying the rotation 
and scaling of the data so that the coordinates become uncorrelated with variance of 1 each.

## Dimensionality Reduction: PCA

Dimensionality reduction is the task of deriving a set of new
artificial features that is smaller than the original feature
set while retaining most of the variance of the original data.
Here we'll use a common but powerful dimensionality reduction
technique called Principal Component Analysis (PCA).
We'll perform PCA on the iris dataset that we saw before:

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
print shape(X),shape(y)
```

PCA is performed using linear combinations of the original features
using a truncated Singular Value Decomposition of the matrix X so
as to project the data onto a base of the top singular vectors.
If the number of retained components is 2 or 3, PCA can be used
to visualize the dataset.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, whiten=True)
pca.fit(X)
```

Once fitted, the pca model exposes the singular vectors in the components_ attribute:

```python
pca.components_
```

Other attributes are available as well:

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

Let us project the iris dataset along those first two dimensions:

```python
X_pca = pca.transform(X)
print shape(X), shape(X_pca)
```

PCA `normalizes` and `whitens` the data, which means that the data
is now centered on both components with unit variance:

```python
import numpy as np
np.round(X_pca.mean(axis=0), decimals=5)
```

```python
np.round(X_pca.std(axis=0), decimals=5)
```

Furthermore, the samples components do no longer carry any linear correlation:

```python
np.corrcoef(X_pca.T)
```

We can visualize the projection using pylab, but first
let's make sure our ipython notebook is in pylab inline mode

```python
%pylab inline
```

Now we can visualize the results using the following utility function:

```python
import pylab as pl
from itertools import cycle

def plot_PCA_2D(data, target, target_names):
    colors = cycle('rgbcmykw')
    target_ids = range(len(target_names))
    pl.figure()
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1],
                   c=c, label=label)
    pl.legend()
```

Now calling this function for our data, we see the plot:

```python
plot_PCA_2D(X_pca, iris.target, iris.target_names)
```

Note that this projection was determined *without* any information about the
labels (represented by the colors): this is the sense in which the learning
is **unsupervised**.  Nevertheless, we see that the projection gives us insight
into the distribution of the different flowers in parameter space: notably,
*iris setosa* is much more distinct than the other two species.

Note also that the default implementation of PCA computes the
singular value decomposition (SVD) of the full
data matrix, which is not scalable when both ``n_samples`` and
``n_features`` are big (more that a few thousands).
If you are interested in a number of components that is much
smaller than both ``n_samples`` and ``n_features``, consider using
`sklearn.decomposition.RandomizedPCA` instead.

Other dimensionality reduction techniques which are useful to know about:

- `sklearn.decomposition.PCA`: <br/>
   Principal Component Analysis
- `sklearn.decomposition.RandomizedPCA`:<br/>
   fast non-exact PCA implementation based on a randomized algorithm
- `sklearn.decomposition.SparsePCA`:<br/>
   PCA variant including L1 penalty for sparsity
- `sklearn.decomposition.FastICA`:<br/>
   Independent Component Analysis
- `sklearn.decomposition.NMF`:<br/>
   non-negative matrix factorization
- `sklearn.manifold.LocallyLinearEmbedding`: <br/>
   nonlinear manifold learning technique based on local neighborhood geometry
- `sklearn.manifold.IsoMap`: <br/>
   nonlinear manifold learning technique based on a sparse graph algorithm

### Solution: Randomized PCA

Load the Iris data:

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
print shape(X),shape(y)
```

The <tt>number_of_components</tt> variable. Change this to observe the explained variance for different number of components!

```python
number_of_components = 2
```

Perform Randomized PCA:

```python
from sklearn.decomposition import RandomizedPCA
pca = RandomizedPCA(n_components=number_of_components, whiten=True)
pca.fit(X)
```

Display the top <tt>number_of_components</tt> eigenvectors:

```python
pca.components_
```

Display the ratio of variances explained by each of the eigenvectors:

```python
pca.explained_variance_ratio_
```

The sum of the variances explained is 1. However, we note that for any value of <tt>number_of_components</tt>, even 1, (can be changed above), the explained variance always sums to 1. Therefore, <tt>explained_variance_ratio</tt> does not indicate the explained variance of the original data, but instead the transformed data.

```python
pca.explained_variance_ratio_.sum()
```

Reduce the dimensions of the data:

```python
X_pca = pca.transform(X)
print shape(X), shape(X_pca)
```

Check that the data has mean = 0:

```python
pca.explained_variance_ratio_
```

```python
np.round(X_pca.mean(axis=0), decimals=5)
```

Check that the data has standard deviation = 1:

```python
np.round(X_pca.std(axis=0), decimals=5)
```

The samples' components do not carry any linear correlation. The values of correlation between different components are effectively 0 (though in actuality a very small value due to rounding errors).

```python
np.corrcoef(X_pca.T)
```

We plot the data to observe the outcome. In this case, we observe that Randomized PCA does not differ significantly from PCA, except is flipped along the X-axis.

```python
if number_of_components == 2:
    plot_PCA_2D(X_pca, iris.target, iris.target_names)
else:
    print "The number of components must be 2 for a 2D graph."
```

The graph generated with <tt>number_of_components = 2</tt> makes it obvious that 1 dimension cannot sufficiently explain the data. However, the <tt>explained_variance_ratio</tt> method still indicates a total explained variance of 1.0.

After investigation, it is found that the <tt>explained_variance_ratio</tt> function only computes the total variance of just the transformed data along the given number of components (2 by default in this notebook) and calculates the ratio of the variance of transformed data along each component with the total. This does not take original data into account and obviously returns a value of 1.0 each time.

Hence, we first use the <tt>var()</tt> method to manually calculate the total variance of the original data:

```python
print "Original variance along each dimension:", X.var(0)
print "Total variance of original data:", X.var(0).sum()
```

Variance explained by each of the top <tt>number_of_components</tt> components for the transformed data:

```python
pca.explained_variance_
```

We can now calculate the ratio of explained variance along each component with respect to the original data:

```python
pca.explained_variance_ / X.var(0).sum()
```

The ratio of total explained variance with respect to variance of the original data is:

```python
sum(pca.explained_variance_ / X.var(0).sum())
```

Due to rounding errors, this value may be slightly lower or higher than the true value, but not by a significant margin.

This gives us the explained variance for any number of components as generated by RandomizedPCA. In the case of this data, it can be observed that PCA and RandomizedPCA perform the same and explain the same amount of variance for the same number of components. For RandomizedPCA, the <tt>number_of_components</tt> variable can be set to the desired number of components for different trials.
